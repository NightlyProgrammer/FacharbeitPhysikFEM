```python
#FYI: dieses Script wurde teilweise Ki generiert, da ähnliche Graphen online erhältlich sind und es sich nur um graphische Darstellung handlet
#die eigentliche FEM implementations wurde keinenfalls mit Ki verfasst

import matplotlib.pyplot as plt
import numpy as np

# Daten definieren
x = np.linspace(0, 9, 10)  # Knoten x0 bis x9
u = np.array([0.8, 0.9, 1.0, 0.95, 0.85, 0.7, 0.6, 0.65, 0.75, 0.9])
offset = 1.5  # Vertikaler Abstand zwischen Basisfunktionen und u(x)

fig, ax = plt.subplots(figsize=(10, 6))

# 1. Hütchenfunktionen (Psi) zeichnen
for i in range(len(x)):
    # Dreieck erstellen: (x[i-1], 0) -> (x[i], 1) -> (x[i+1], 0)
    x_tri = [x[max(0, i-1)], x[i], x[min(len(x)-1, i+1)]]
    y_tri = [0, 1, 0]
    ax.plot(x_tri, y_tri, color='black', lw=1)
    
    # Spezielle Einfärbung für Psi_5 und Psi_6 (wie im Bild)
    if i == 5:
        ax.fill(x_tri, y_tri, color='palevioletred', alpha=0.5)
    elif i == 6:
        ax.fill(x_tri, y_tri, color='skyblue', alpha=0.7)

# 2. Obere Kurve (Interpolation u)
u_plot = u + offset
ax.plot(x, u_plot, 'r-', lw=2)  # Rote Verbindungslinie
ax.scatter(x, u_plot, color='black', s=30, zorder=5) # Diskrete Werte u_i

# 3. GESTRICHELTE VERBINDUNGSLINIEN (Korrelation)
# Zeichnet Linien von der Spitze jeder Psi_i (y=1) zum Wert u_i
ax.vlines(x, 1, u_plot, colors='gray', linestyles='dashed', lw=0.8, alpha=0.6)

# 4. Optische Anpassungen
ax.set_xticks(x)
ax.set_xticklabels([f'$x_{i}$' for i in range(len(x))])
ax.set_yticks([1])
ax.set_yticklabels(['1'])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel('$x$')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Daten definieren
num = 8
x = np.linspace(0, 9, 10)[:num]
u = np.array([0.8, 0.9, 1.0, 0.95, 0.85, 0.7, 0.6, 0.65, 0.75, 0.9])[:num]
offset = 1.5

fig, ax = plt.subplots(figsize=(12, 6))

for i in range(len(x)):
    # 1. Hütchenfunktionen zeichnen
    x_tri = [x[max(0, i-1)], x[i], x[min(len(x)-1, i+1)]]
    y_tri = [0, 1, 0]
    ax.plot(x_tri, y_tri, color='black', lw=1)
    
    # Farben für Psi_5 und Psi_6
    if i == 5: ax.fill(x_tri, y_tri, color='palevioletred', alpha=0.5)
    elif i == 6: ax.fill(x_tri, y_tri, color='skyblue', alpha=0.7)

    # 2. Beschriftungen hinzufügen (LaTeX)
    # Psi_i an der Spitze der Hütchen (y=1)
    ax.text(x[i], 1.05, f'$\Phi_{{{i}}}$', ha='center', fontsize=12)
    
    # u_i über den Datenpunkten (y = u + offset)
    ax.text(x[i], u[i] + offset + 0.05, f'$u_{{{i}}}$', ha='center', fontsize=12)

# 3. Obere Kurve und Verbindungslinien
u_plot = u + offset
ax.plot(x, u_plot, 'r-', lw=2, zorder=4)
ax.scatter(x, u_plot, color='black', s=30, zorder=5)
ax.vlines(x, 1, u_plot, colors='gray', linestyles='dashed', lw=0.8, alpha=0.6)

# Optik finalisieren
ax.set_xticks(x)
ax.set_xticklabels([f'$x_{{{i}}}$' for i in range(len(x))])
ax.set_yticks([1])
ax.set_yticklabels(['1'])
ax.set_ylim(0, 3) # Platz für Beschriftungen schaffen
ax.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
```
