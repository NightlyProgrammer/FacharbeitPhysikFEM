```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Dieses Script wurde benutzt um den Graphen der benutzten Ansatz-Funktion zu erstellen
#Dieses Script muss nicht für das Verständnis der Facharbeit gelesen werden
from abc import ABC, abstractmethod
#Define Abstract Base Class
class ShapeFuntion(ABC):
    all_shape_functions = []#array of all shape function 
    def __init_subclass__(cls, **kwargs):#this code snippet just makes the code cleaner, its not that important to understand for the Facharbeit
        super().__init_subclass__(**kwargs)
        # Automatically adds any child class to the list
        ShapeFuntion.all_shape_functions.append(cls())

    @abstractmethod
    def N0(self, psi): pass
    @abstractmethod
    def N1(self, psi): pass
    @property
    @abstractmethod
    def function_name(self): pass
```

```python
class LinearShapeFunction(ShapeFuntion):
    function_name = "Lineare Ansatzfunktion"
    def N0(self, psi):
        return 0.5 - 0.5 * psi
    def N1(self, psi):
        return 0.5 + 0.5 * psi
```

```python
x_axis = np.linspace(-1, 1, 100)#x axis in interval [-1, 1]
for function in ShapeFuntion.all_shape_functions:
    plt.close()
    y_axis_n0 = function.N0(x_axis)
    y_axis_n1 = function.N1(x_axis)

    plt.plot(x_axis, y_axis_n0, label="N0")
    plt.plot(x_axis, y_axis_n1, label="N1")
    plt.legend()
    plt.xlabel(function.function_name)
    plt.show()
```
